# 3.2 Data Cleaning and Summary Statistics [3]

* Data Cleaning

1. If found any missing values use appropriate data imputations techniques discussed in your work-
shops to fill the missing values. Please justify why did you pick particular techniques.

Choosing the median for filling in missing values makes sense because it doesn't get messed up by really high or low numbers. It's like picking the middle value, which is helpful when there are strange numbers in the data. Also, the median works well with all kinds of data shapes. So, it's like a safe bet that works for different types of information in our dataset.

```python
#Q1-Code:

for column in sk.columns:
    # Check if the column has missing values
    if sk[column].isnull().any():
        # Calculate the median of the column
        median_value = sk[column].median()

        # Fill missing values with the median
        sk[column].fillna(median_value)

# Display the DataFrame with missing values filled using median
print(sk)

```

```python
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = '/content/drive/MyDrive/AI-Assignment1/WineQT.csv'  # Replace with the actual file path
data = pd.read_csv(file_path)

# Filter the 'quality' column by values 3, 4, 5, 6, 7, 8
selected_quality_values = [3, 4, 5, 6, 7, 8]
filtered_quality_values = data[data['quality'].isin(selected_quality_values)]['quality']

# Sort the filtered quality values
sorted_quality_values = filtered_quality_values.sort_values()

# Adjust figure size
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the bar chart with color and grid
bars = ax.bar(sorted_quality_values.value_counts().index, sorted_quality_values.value_counts().values, color='skyblue', edgecolor='black')

# Add labels and title
ax.set_xlabel('Wine Quality')
ax.set_ylabel('Quantity')
ax.set_title('Distribution of Wine Quality')

# Add grid
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Add data labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 1), ha='center', va='bottom', color='black', fontsize=8)

# Add a summary with count values
summary = f"Count of each quality value:\n{sorted_quality_values.value_counts()}"
plt.figtext(0.1, -0.15, summary, wrap=True, horizontalalignment='left', fontsize=10)

plt.savefig('/content/drive/MyDrive/AI-Assignment1/WineQualityDistribution.png')

plt.show()

```

```python

# Importing Seaborn and Matplotlib for data visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Create a correlation matrix
corr_matrix = sk.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using Seaborn
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# Adding title to the plot
plt.title("Correlation Matrix", fontsize=16)

# Adjusting layout for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.subplots_adjust(bottom=0.15, left=0.15)  # Adjusting the bottom and left margins

# Saving the heatmap as an image with higher resolution
plt.savefig('/content/drive/MyDrive/AI-Assignment1/heatmap_beautiful.png', dpi=300)

# Displaying the plot
plt.show()

```
